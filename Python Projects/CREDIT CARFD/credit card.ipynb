```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to panadas dataframe
credit_card_data=pd.read_csv('creditcard.csv')
print(credit_card_data)
```

```python
# dataset information: it tells about the information of datas
credit_card_data.info()
```

```python
# checking the nmber of missing values in each column
credit_card_data.isnull().sum()
```

```python
# DISTRIBUTION OF LEGIT TRANSACTION & FRAUDLENT TYRANSACTION
credit_card_data['Class'].value_counts()
```

```python
# SEPARATING THE DATA FOR ANALYSIS
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
# Build a sample dataset containing similar distribution of normal transaction and fraudulent Transactions
# Number of Fraudulent Transactions-->492

legit_sample=legit.sample(n=492) 
# -->this will take random 492 data set from legit data set...
# CONCADINATE 2 DATA FRAMES
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset
# HERE WE HAVE CONCADINATED LEGIT SAMPLE JETA RANDOMLY KOTA NORMAL LEGIT TRANSACTION NEBE 492 MOTO AAR TAR SATHE FRAUD ER J 492 DATA ACHE SETAA ADD HOLO..BUT SETA AS EKHANE AXIS =0 SO LEGIT SAMPLE COLUMN   ER NICHE ROW WISE ADD HBE..AR AXIS=1 HOLE COLUMN WISE  AD HBE
```

```python
# Splitting the data into FEATURES & TARGETS 
X=new_dataset.drop(columns='Class',axis=1)

Y=new_dataset['Class']
print(X)
print(Y)
```

```python
# SPLIT THE DATA INTO tRAINING DATA & TESTING DATA
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# MODEL TRAING
# lOGISTIC REGRESS
model=LogisticRegression()
model.fit(X_train,Y_train)
 
```

```python
# MODEL EVALUATION
#  ACCURACY SCORE
#  ACCURACY SCORE ON TRAINING DATA
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy on Traing data :',training_data_accuracy*100,"%")
```

```python
#  ACCURACY SCORE ON Test DATA
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy on Test data :',test_data_accuracy*100,"%")
```
